#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

```python
# TODO
import pandas as pd

df = pd.read_csv("assets/real_estate.csv", sep=";")

# Wow! another time (I copied from Exercise 14)
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
south_belt_df = df[df['level5'].isin(south_belt_populations)]

# Create a new column for price per square meter
south_belt_df['pps'] = south_belt_df['price'] / south_belt_df['surface']

# Filter the DataFrame for Getafe and Alcorcón
getafe_alcorcon_df = south_belt_df[south_belt_df['level5'].isin(['Getafe', 'Alcorcón'])]

# Calculate the average price per square meter for both towns
getafe_pps = getafe_alcorcon_df[getafe_alcorcon_df['level5'] == 'Getafe']['pps'].mean()
alcorcon_pps = getafe_alcorcon_df[getafe_alcorcon_df['level5'] == 'Alcorcón']['pps'].mean()

# Print the results
print(f"The average price per square meter in Getafe is {getafe_pps:.2f} USD/m^2.")
print(f"The average price per square meter in Alcorcón is {alcorcon_pps:.2f} USD/mm^2.")

# Optional: Compare the values
if getafe_pps > alcorcon_pps:
    print("Getafe is more expensive per square meter than Alcorcón.")
else:
    print("Alcorcón is more expensive per square meter than Getafe.")
```

```python
from ipyleaflet import Map, Marker, Popup, basemaps
from ipywidgets import HTML

# Okay, I thought the one before this was advanced. . . this is crazy! This is just the prework! I feel bad for looking up how to do this. I reread the readings for each of the languages and still couldn't do it. Sorry for using ChatGPT. I am going to go read some books on this.
# Create a map centered on Madrid (default location for demonstration)
map = Map(center=(40.4, -3.7), zoom=11, min_zoom=1, max_zoom=20, basemap=basemaps.OpenStreetMap.Mapnik)

# Filter the data for the south belt of Madrid (subset from Exercise 14)
south_belt_df = df[df['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])]

# Create a dictionary to assign different colors to each population
color_map = {
    'Fuenlabrada': 'red',
    'Leganés': 'blue',
    'Getafe': 'green',
    'Alcorcón': 'orange'
}

# Create markers for each real estate in the south belt
markers = []
for idx, row in south_belt_df.iterrows():
    latitude = row['latitude']
    longitude = row['longitude']
    population = row['level5']
    
    # Create a marker for the estate
    marker = Marker(location=(latitude, longitude), color=color_map[population])
    
    # Add a popup to the marker with information about the house
    popup_content = f"Address: {row['realEstate_name']}<br>Price: {row['price']} USD<br>Population: {population}"
    popup = Popup(child=HTML(value=popup_content))
    
    # Set the popup on the marker
    marker.popup = popup
    
    # Append marker to the list of markers
    markers.append(marker)

# Add all markers to the map
for marker in markers:
    map.add_layer(marker)

# Display the map
map



```
