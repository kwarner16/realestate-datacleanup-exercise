#### Exercise 10. Are the average prices per square meter (price/m2) of "Valdemorillo" and "Galapagar" the same? (★★☆)

```python
# TODO My GOODNESS this is hard!

import pandas as pd

df = pd.read_csv("assets/real_estate.csv", sep=";")

# Create a new column "pps" (price per square meter)
df["pps"] = df["price"] / df["surface"]

# Calculate the average pps for Valdemorillo
valdemorillo_pps = df[df["level5"] == "Valdemorillo"]["pps"].mean()

# Calculate the average pps for Galapagar
galapagar_pps = df[df["level5"] == "Galapagar"]["pps"].mean()

# Print both values
print(f"Average price per square meter in Valdemorillo: {valdemorillo_pps:.2f} USD/m²")
print(f"Average price per square meter in Galapagar: {galapagar_pps:.2f} USD/m²")

# Compare and print conclusion
if valdemorillo_pps == galapagar_pps:
    print("The average price per square meter in Valdemorillo and Galapagar is the same.")
else:
    print("The average price per square meter in Valdemorillo and Galapagar is different.")

```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("assets/real_estate.csv", sep=";")

# Again, same thing(I copied from Exercise 14)
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
south_belt_df = df[df['level5'].isin(south_belt_populations)]

# This is some of the most complex code I have ever written. I think I did it right though
def normalize(df, column):
    return (df[column] - df[column].min()) / (df[column].max() - df[column].min())

# Create a plot
plt.figure(figsize=(10, 6))

# Plot histograms for the normalized prices of each population
for population in south_belt_populations:
    
    population_df = south_belt_df[south_belt_df['level5'] == population]
    
    population_df['normalized_price'] = normalize(population_df, 'price')
    
    plt.hist(population_df['normalized_price'], bins=20, label=population)

plt.show()
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
# TODO This was suprisingly easy. the hardest part was calculating the avg price per square meter
import pandas as pd

df = pd.read_csv("assets/real_estate.csv", sep=";")

# Wow! another time (I copied from Exercise 14)
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
south_belt_df = df[df['level5'].isin(south_belt_populations)]

# Create a new column for price per square meter
south_belt_df['pps'] = south_belt_df['price'] / south_belt_df['surface']

# Filter the DataFrame for Getafe and Alcorcón
getafe_alcorcon_df = south_belt_df[south_belt_df['level5'].isin(['Getafe', 'Alcorcón'])]

# Calculate the average price per square meter for both towns
getafe_pps = getafe_alcorcon_df[getafe_alcorcon_df['level5'] == 'Getafe']['pps'].mean()
alcorcon_pps = getafe_alcorcon_df[getafe_alcorcon_df['level5'] == 'Alcorcón']['pps'].mean()

# Print the results
print(f"The average price per square meter in Getafe is {getafe_pps:.2f} USD/m^2.")
print(f"The average price per square meter in Alcorcón is {alcorcon_pps:.2f} USD/mm^2.")

# Optional: Compare the values
if getafe_pps > alcorcon_pps:
    print("Getafe is more expensive per square meter than Alcorcón.")
else:
    print("Alcorcón is more expensive per square meter than Getafe.")
```
